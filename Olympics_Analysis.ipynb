```python
def gold_medal(olympics_merged)  :
    if olympics_merged["Medal"] == "Gold" :
        return 1 
    else :
        return 0
def silver_medal(olympics_merged)  :
    if olympics_merged["Medal"] == "Silver" :
        return 1 
    else :
        return 0
def bronze_medal(olympics_merged)  :
    if olympics_merged["Medal"] == "Bronze" :
        return 1 
    else :
        return 0
    
olympics_merged["Gold"]   = olympics_merged.apply(lambda olympics_merged:gold_medal(olympics_merged),axis = 1) 
olympics_merged["Silver"] = olympics_merged.apply(lambda olympics_merged:silver_medal(olympics_merged),axis = 1) 
olympics_merged["Bronze"] = olympics_merged.apply(lambda olympics_merged:bronze_medal(olympics_merged),axis = 1) 
olympics_merged["Total Medals"] = olympics_merged["Gold"] + olympics_merged["Silver"] +olympics_merged["Bronze"]
```

# **Change of Number of events in an olympics over the years**

```python
olympics_merged_summer = olympics_merged[olympics_merged['Season'] == 'Summer']
olympics_merged_winter = olympics_merged[olympics_merged['Season'] == 'Winter']
year = olympics_merged_summer['Year'].sort_values()
year = year.unique()
counts = []

for item in year:
    counter = list(olympics_merged_summer[olympics_merged_summer['Year'] == item]['Sport'].unique())
    counts.append(len(counter))
    
year1 = olympics_merged_winter['Year'].sort_values()
year1 = year1.unique()
counts1 = []

for item in year1:
    counter1 = list(olympics_merged_winter[olympics_merged_winter['Year'] == item]['Sport'].unique())
    counts1.append(len(counter1))        
```

```python
from matplotlib import style

style.use("ggplot")

plt.figure(figsize=(15, 13))

# Summer Olympics
plt.subplot(211)
plt.xticks(rotation=60)
sns.pointplot(x=year, y=counts, markers="o", color="blue")
plt.title("Sports over the last 12 decades in Summer Olympics", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Total Number of Sports")

# Winter Olympics
plt.subplot(212)
plt.xticks(rotation=60)
sns.pointplot(x=year1, y=counts1, markers="o", color="red")
plt.title("Sports over the last 12 decades in Winter Olympics", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Total Number of Sports")

plt.tight_layout()  
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Check for missing values
df.isnull().sum()

# Fill missing Age, Height, and Weight with median values
df["Age"].fillna(df["Age"].median(), inplace=True)
df["Height"].fillna(df["Height"].median(), inplace=True)
df["Weight"].fillna(df["Weight"].median(), inplace=True)

# Fill missing Medal column with "No Medal"
df["Medal"].fillna("No Medal", inplace=True)

# Drop unnecessary columns
df.drop(columns=["notes"], inplace=True)

# Check for duplicate rows
df.drop_duplicates(inplace=True)

# Convert categorical variables to lowercase for consistency
df["Sex"] = df["Sex"].str.lower()
df["Team"] = df["Team"].str.lower()
df["Sport"] = df["Sport"].str.lower()
df["Event"] = df["Event"].str.lower()
```

```python
df
```

```python
numeric_df = df.select_dtypes(include=['number'])

# Create a new column for total medals won per athlete
df["Total Medals"] = df[["Gold", "Silver", "Bronze"]].sum(axis=1)

# Create a binary column to indicate if an athlete won any medal
df["Won Medal"] = np.where(df["Total Medals"] > 0, 1, 0)

# Encode "Sex" column (Male=1, Female=0)
df["Sex"] = df["Sex"].map({"male": 1, "female": 0})

# Calculate BMI for each athlete
df["BMI"] = df["Weight"] / ((df["Height"] / 100) ** 2)

# Check correlation between features
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
df_encoded = df.copy()

# Convert categorical columns to numeric using Label Encoding
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    label_encoders[col] = LabelEncoder()
    df_encoded[col] = label_encoders[col].fit_transform(df[col])

# Generate the heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df_encoded.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
top_sports = df.groupby("Sport")["Total Medals"].sum().sort_values(ascending=False).head(5)
print(top_sports)
```

```python
# Ensure 'Sex' and 'Total Medals' exist and have valid data
if "Sex" in df.columns and "Total Medals" in df.columns and not df.empty:
    df = df.dropna(subset=["Sex", "Total Medals"])  # Remove NaNs
    gender_medals = df.groupby("Sex")["Total Medals"].sum()

    if not gender_medals.empty:  # Ensure there's data to plot
        colors = ["red", "blue"] if len(gender_medals) > 1 else ["red"]
        gender_medals.plot(kind="bar", color=colors)
        plt.title("Total Medals by Gender")
        plt.xticks(rotation=0)
        plt.show()
    else:
        print("No valid data to plot.")
else:
    print("Columns 'Sex' or 'Total Medals' not found in DataFrame.")
```

```python
# Grouping and calculating total medals
olympics_yearly = olympics_merged.groupby('Year')[['Gold', 'Silver', 'Bronze']].sum().reset_index()
olympics_yearly['Total Medals'] = olympics_yearly['Gold'] + olympics_yearly['Silver'] + olympics_yearly['Bronze']

# Plot using Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=olympics_yearly, x="Year", y="Total Medals", marker="o", color="blue")

# Customizing the plot
plt.title("Olympic Medals Trend Over Time")
plt.xlabel("Year")
plt.ylabel("Total Medals")
plt.grid(True)
plt.show()
```

```python
sns.barplot(data=olympics_yearly, x="Year", y="Total Medals", color="blue")
plt.xticks(rotation=45)
plt.title("Total Olympic Medals Growth Over Time")
plt.xlabel("Year")
plt.ylabel("Total Medals")
plt.show()
```

```python
olympics_yearly['Medal_MA'] = olympics_yearly['Total Medals'].rolling(window=5).mean()
plt.figure(figsize=(12,6))
plt.plot(olympics_yearly['Year'], olympics_yearly['Total Medals'], label="Total Medals", marker="o")
plt.plot(olympics_yearly['Year'], olympics_yearly['Medal_MA'], label="5-Year Moving Average", linestyle="--", color="red")
plt.title("Olympic Medals Trend with Moving Average")
plt.xlabel("Year")
plt.ylabel("Total Medals")
plt.legend()
plt.show()
```

```python
top_countries = olympics_merged.groupby('Team')['Total Medals'].sum().nlargest(5).index
olympics_top = olympics_merged[olympics_merged['Team'].isin(top_countries)]
plt.figure(figsize=(15, 6))
sns.barplot(data=olympics_top, x="Year", y="Total Medals", hue="Team")
plt.title("Total Medals by Top 5 Countries Over Time")
plt.xticks(rotation=45)
plt.show()
```

```python
pivot_df = olympics_merged.pivot_table(index="Sport", columns="Year", values="Total Medals", aggfunc="sum")
plt.figure(figsize=(15, 8))
sns.heatmap(pivot_df, cmap="coolwarm", linewidths=0.5)
plt.title("Heatmap of Medals per Sport Over Time")
plt.show()
```
